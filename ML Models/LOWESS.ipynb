*Implement the non-parametric locally weighted regression algorithm in order to fit data points.*

# Required Modules:

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

# Generating Our Own Random Dataset:

```python
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])
```

# Sorting the data for plotting:

```python
sorted_indices = X.flatten().argsort()
X_sorted = X[sorted_indices]
y_sorted = y[sorted_indices]
```

# Lowess (Locally Weighted Scatterplot Smoothing) Statsmodel:

```python
lowess = sm.nonparametric.lowess(y_sorted, X_sorted.flatten(), frac=0.3)
```

# Graph:

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='black', label='Data points')
plt.plot(lowess[:, 0], lowess[:, 1], color='red', label='LOWESS')
```
