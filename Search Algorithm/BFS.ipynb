# Required Modules:

```python
from collections import deque
import networkx as nx
import matplotlib.pyplot as plt
```

# Simple Graph Implementation:

```python
class Graph:
    
    def __init__(self):
        self.graph = {}

    def add_edge(self, node, neighbor):
        if node not in self.graph:
            self.graph[node] = []
        self.graph[node].append(neighbor)
```

# Breadth First Search Algorithm:

```python
def breadth_first_search(graph, start, goal):
    visited = set()
    queue = deque([(start, [start])])

    while queue:
        current, path = queue.popleft()

        if current in visited:
            continue

        visited.add(current)

        if current == goal:
            return path

        for neighbor in graph.graph.get(current, []):
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor]))

    return None
```

# Example 1 :

```python
graph = Graph()
graph.add_edge('A', 'B')
graph.add_edge('A', 'C')
graph.add_edge('B', 'D')
graph.add_edge('B', 'E')
graph.add_edge('C', 'F')
```

```python
G = nx.Graph(graph.graph)
pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=10, font_color='black', edge_color='gray')
plt.title('Graph Visualization')
plt.show()
```

```python
start_node = 'A'
goal_node = 'F'
result = breadth_first_search(graph, start_node, goal_node)
```

```python
if result:
    print(f"Breadth-First Search path from {start_node} to {goal_node}: {result}")
else:
    print(f"No path found from {start_node} to {goal_node}")
```
