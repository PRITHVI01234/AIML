# Required Modules:

```python
from collections import deque
import networkx as nx
import matplotlib.pyplot as plt
```

# Simple Graph Implementation:

```python
class Graph:
    
    def __init__(self):
        self.graph = {}

    def add_edge(self, node, neighbor):
        if node not in self.graph:
            self.graph[node] = []
        self.graph[node].append(neighbor)
```

# Depth First Search Algorithm:

```python
def depth_first_search(graph, current, goal, visited, path):
    if current == goal:
        return path + [current]

    if current in visited:
        return None

    visited.add(current)

    for neighbor in graph.graph.get(current, []):
        new_path = depth_first_search(graph, neighbor, goal, visited, path + [current])
        if new_path:
            return new_path

    return None
```

# Example 1 :

```python
graph = Graph()
graph.add_edge('A', 'B')
graph.add_edge('A', 'C')
graph.add_edge('B', 'D')
graph.add_edge('B', 'E')
graph.add_edge('C', 'F')
```

```python
G = nx.Graph(graph.graph)
pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=10, font_color='black', edge_color='gray')
plt.title('Graph Visualization')
plt.show()
```

```python
start_node = 'A'
goal_node = 'F'
visited_nodes = set()
result = depth_first_search(graph, start_node, goal_node, visited_nodes, [])
```

```python
if result:
    print(f"Depth-First Search path from {start_node} to {goal_node}: {result}")
else:
    print(f"No path found from {start_node} to {goal_node}")
```
